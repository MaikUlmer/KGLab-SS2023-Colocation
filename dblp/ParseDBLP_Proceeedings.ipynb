```python
import requests
import  bs4
import pandas as pd
from bs4 import Tag
from bs4 import BeautifulSoup as soup
```

```python
result000_3499 = {"ProceedingsTitle": [],
          "Proceedings": [], 
          "Vol_Number": [],
          "Year": [],  
          "persistent_url": []  
          }
```

```python

for i in range (35):
    url = 'https://dblp.org/db/series/ceurws/ceurws'+str(i)+'00-'+str(i)+'99'
    r= requests.get(url)
    html = bs4.BeautifulSoup(r.text, "html.parser")
    all_proceedings = html.find("ul", class_="publ-list").find_all("li")
    for data in all_proceedings:
        ProceedingsTitle = data.find("span", class_="title")
        Proceedings = data.find("span", itemprop="isPartOf")
        Vol_Number = data.find("div", class_="nr")
        Year = data.find("span", itemprop="datePublished")
        persistent_url = data.find("a", class_="toc-link")
        
        result000_3499["ProceedingsTitle"].append(ProceedingsTitle)
        result000_3499["Proceedings"].append(Proceedings)
        result000_3499["Vol_Number"].append(Vol_Number)
        result000_3499["Year"].append(Year)
        result000_3499["persistent_url"].append(persistent_url)
```

```python
df000_3499 = pd.DataFrame(result000_3499)
df000_3499.head()
```

```python
df000_3499_dropna=df000_3499.dropna(axis=0, how='any')
df000_3499_dropna.head()
```

```python
df000_3499_dropna['ProceedingsTitle'] = df000_3499_dropna['ProceedingsTitle'].astype(str)
df000_3499_dropna['Proceedings'] = df000_3499_dropna['Proceedings'].astype(str)
df000_3499_dropna['Vol_Number'] = df000_3499_dropna['Vol_Number'].astype(str)
df000_3499_dropna['Year'] = df000_3499_dropna['Year'].astype(str)
df000_3499_dropna['persistent_url'] = df000_3499_dropna['persistent_url'].astype(str)
```

```python
df000_3499_dropna['ProceedingsTitle'] = df000_3499_dropna['ProceedingsTitle'].str.replace(r'<span class="title" itemprop="name">', '') 
df000_3499_dropna['ProceedingsTitle'] = df000_3499_dropna['ProceedingsTitle'].str.replace(r'</span>', '') 
df000_3499_dropna['Proceedings'] = df000_3499_dropna['Proceedings'].str.replace(r'<span itemprop="isPartOf" itemscope="" itemtype="http://schema.org/BookSeries"><span itemprop="name">', '') 
df000_3499_dropna['Proceedings'] = df000_3499_dropna['Proceedings'].str.replace(r'</span></span>', '') 
df000_3499_dropna['Vol_Number'] = df000_3499_dropna['Vol_Number'].str.replace(r'<div class="nr">', '') 
df000_3499_dropna['Vol_Number'] = df000_3499_dropna['Vol_Number'].str.replace(r'</div>', '') 
df000_3499_dropna['Year'] = df000_3499_dropna['Year'].str.replace(r'<span itemprop="datePublished">', '') 
df000_3499_dropna['Year'] = df000_3499_dropna['Year'].str.replace(r'</span>', '') 
df000_3499_dropna['persistent_url'] = df000_3499_dropna['persistent_url'].str.replace(r'<a class="toc-link" href="', '') 
df000_3499_dropna['persistent_url'] = df000_3499_dropna['persistent_url'].str.replace(r'">[contents]</a>', '') 
```

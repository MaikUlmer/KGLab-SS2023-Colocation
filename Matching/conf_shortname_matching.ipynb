```python
volums = pd.read_csv('ceur_wiki_dblp.csv')
conf= pd.read_csv('conf_shortname.csv')
```

```python
no_coloc=volums[volums['coloc_conf'].isnull()]
```

```python
conf['PAKM'] = conf['PAKM'].astype(str)
conf_shortname_list=conf['PAKM'].tolist()
no_coloc['ceur_long_title'] = no_coloc['ceur_long_title'].astype(str)
no_coloc['volname'] = no_coloc['volname'].astype(str)
no_coloc['coloc_conf'] = no_coloc['coloc_conf'].astype(str)
no_coloc['voltitle'] = no_coloc['voltitle'].astype(str)
no_coloc['h1'] = no_coloc['h1'].astype(str)
no_coloc['tdtitle'] = no_coloc['tdtitle'].astype(str)
no_coloc['acronym'] = no_coloc['acronym'].astype(str)
no_coloc['title'] = no_coloc['title'].astype(str)
no_coloc['h1'] = no_coloc['h1'].astype(str)
no_coloc['workshopLabel'] = no_coloc['workshopLabel'].astype(str)
no_coloc['ProceedingsTitle'] = no_coloc['ProceedingsTitle'].astype(str)
no_coloc['ShortName'] = no_coloc['ShortName'].astype(str)
```

```python
def filter_and_add_matching_words(df, title_columns, string_list):
    # Filter rows based on whether at least one specified column contains strings from the list
    filtered_rows = df[df[title_columns].apply(lambda x: any(string in val for string in string_list for val in x), axis=1)].copy()

    # Add a new column to show which words from the list are contained in each column
    for col in title_columns:
        filtered_rows[col + '_contained_words'] = filtered_rows[col].apply(
            lambda title: ', '.join([string for string in string_list if string in title])
        )

    return filtered_rows
```

```python
# Specify the list of column names and the string list
title_columns = ['ceur_long_title', 'tdtitle','volname','acronym','voltitle','h1','title','workshopLabel','ShortName']
string_list = conf_shortname_list

# Apply the function to filter and add the contained words
result = filter_and_add_matching_words(no_coloc, title_columns, string_list)

print(result)
```

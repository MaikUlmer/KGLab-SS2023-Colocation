```python
import numpy as np
import pandas as pd
import re
```

```python
ceur = pd.read_csv('ceur_cleaned_alltitle.csv')
wiki = pd.read_csv('workshop_wiki.csv')
dblp = pd.read_csv('dblp_proceedings0_34XX.csv')
```

```python
# Perform a left join
ceur_wiki = pd.merge(ceur, wiki, left_on='vol_url', right_on='ref_url', how='left')
ceur_wiki.head()
```

```python
# Perform a left join
ceur_wiki_dblp = pd.merge(ceur_wiki, dblp, left_on='vol_number', right_on='Vol_Number', how='left')
ceur_wiki_dblp.head()
```
